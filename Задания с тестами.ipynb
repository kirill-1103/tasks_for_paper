### Задание №1
Ниже представлен код с функцией func, Вам необходимо выбрать из вариантов ответа - для чего нужна эта функция.

```python
from PIL import Image, ImageDraw

def func(img):
    for y in range(img.height):
        for x in range(img.width):
            pixel = img.getpixel((x,y))
            if pixel == (255,0,0):
                img.putpixel((x,y),(0,0,255))
    return img
```

2)Функция заменяет синие пиксели на зеленые

3)**Функция заменяет красные пиксели на синие**

4)Функция закрашивает изображение красным цветом

```python
img = Image.open("colors.png")
modification_img = func(img)
modification_img
```

```python
import cv2

def func(img):
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            (b, g, r) = img[y,x]
            if b == 255 and g ==0 and r == 0:
                img[y,x] = (0, 255, 0)
    return img
```

### Задание №6
Ниже представлен код с функцией func, Вам необходимо выбрать из вариантов ответа - для чего нужна эта функция.

```python
from PIL import Image

def func(img):
    new_img = Image.new(size = (img.width*2, img.height), mode = 'RGB')
    transposed_img = img.transpose(Image.FLIP_LEFT_RIGHT)
    new_img.paste(img,(0,0))
    new_img.paste(transposed_img,(img.width,0))
    return new_img
```

```python
img = Image.open("lion.png")
modification_img = func(img)
modification_img
```

### Задание №7
Ниже представлен код с функцией func, Вам необходимо выбрать из вариантов ответа - для чего нужна эта функция.

```python
import cv2
import numpy as np

def func(img):
    new_img = np.full((img.shape[0], img.shape[1]*2, 3),(255,255,255),dtype=np.uint8)
    fliped_img = cv2.flip(img,flipCode = 1)
    new_img[0:img.shape[0],0:img.shape[1]] = img
    new_img[0:img.shape[0], img.shape[1]:new_img.shape[1]] = fliped_img
    return new_img
```

Варианты ответов:

1)Функция отражает изображение по горизонтальной оси

2)**Функция возвращает изображение, содержащее исходное и отраженное по вертикальной оси**

3)Функция возвращает изображение, содержащее исходное и отраженное по горизонтальной оси

4)Функция отражает изображение по вертикальной оси

```python
img = cv2.imread('lion.png')
img = func(img)
cv2.imshow("image",img)
cv2.waitKey(0)
```

### Задание №8
Ниже представлен код с функцией func, Вам необходимо выбрать из вариантов ответа - для чего нужна эта функция.

```python
from PIL import Image, ImageDraw

def func(img, w, c):
    draw = ImageDraw.Draw(img)
    draw.line((0, 0, img.width, img.height), width = w, fill=c)
    draw.line((img.width, 0, 0, img.height), width = w, fill=c)
    return img
```

Варианты ответов:

1)Функция не изменяет изображение

2)Функция рисует линию на изображении

3)Функция рисует квадрат на изображении

4)**Функция рисует крест на изображении**

```python
img = Image.open("jpg.jpg")
modification_img = func(img,4,(255,211,11))
modification_img
```

### Задание №9
Ниже представлен код с функцией func, Вам необходимо выбрать из вариантов ответа - для чего нужна эта функция.

```python
import cv2

def func(img,w,c):
    cv2.line(img, (0,0), (img.shape[1], img.shape[0]), color = c, thickness = w)
    cv2.line(img, (img.shape[1], 0), (0,img.shape[0]), color = c, thickness = w)
    return img
    
```

Варианты ответов:

1)Функция не изменяет изображение

2)Функция рисует линию на изображении

3)Функция рисует квадрат на изображении

4)**Функция рисует крест на изображении**
